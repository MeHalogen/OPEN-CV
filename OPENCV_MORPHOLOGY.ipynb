```python
def plt_imshow(title, image):
  # convert the image frame BGR to RGB color space and display it
  image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
  plt.imshow(image)
  plt.title(title)
  plt.grid(False)
  plt.show()
```

```python
# import the necessary packages
from matplotlib import pyplot as plt
import argparse
import cv2
import numpy as np
```

```python
args = {
	"image": "scenery.jpg",
  "image1": "OPENCV2.png",
  "noisy_image": "OPENCV_NOISE.png",
  "noisy_image2": "OPENCV_NOISE2.png",
  "car": "car.jpg",
}
```

# EROSION

```python
image = cv2.imread(args["image1"])
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt_imshow("Original", image)

for i in range(0, 3):
  eroded = cv2.erode(gray.copy(), None, iterations=i + 1)
  plt_imshow("Eroded {} times".format(i+1), eroded)
```

# DILATION

```python
plt_imshow("Original", image)

for i in range(0, 7):
  dilated = cv2.dilate(gray.copy(), None, iterations=i+1)
  plt_imshow("Dilated {} times".format(i+1), dilated)
```

# Opening

```python
image_n = cv2.imread(args["noisy_image2"])
gray = cv2.cvtColor(image_n, cv2.COLOR_BGR2GRAY)
plt_imshow("Original", image_n)
kernelSizes = [(3,3), (5,5), (7,7), (9,9)]

for kernelSize in kernelSizes:
  kernel = cv2.getStructuringElement(cv2.MORPH_RECT, kernelSize)
  opening = cv2.morphologyEx(gray, cv2.MORPH_OPEN, kernel)
  plt_imshow("Opening: ({}. {})".format(kernelSize[0], kernelSize[1]), opening)
```

# CLOSING

```python
image_n = cv2.imread(args["image1"])
gray = cv2.cvtColor(image_n, cv2.COLOR_BGR2GRAY)
plt_imshow("Original", image_n)
kernelSizes = [(3,3), (5,5), (7,7), (9,9), (11,11), (13,13), (15,15), (17,17)]

for kernelSize in kernelSizes:
  kernel = cv2.getStructuringElement(cv2.MORPH_RECT, kernelSize)
  closing = cv2.morphologyEx(gray, cv2.MORPH_CLOSE, kernel)
  plt_imshow("closing: ({}. {})".format(kernelSize[0], kernelSize[1]), closing)
```

# MORPHOLOGICAL GRADIENT

```python
image = cv2.imread(args["image1"])
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt_imshow("Original", image)
kernelSizes = [(3,3), (5,5), (7,7), (9,9)]

for kernelSize in kernelSizes:
  kernel = cv2.getStructuringElement(cv2.MORPH_RECT, kernelSize)
  gradient = cv2.morphologyEx(gray, cv2.MORPH_GRADIENT, kernel)
  plt_imshow("Opening: ({}. {})".format(kernelSize[0], kernelSize[1]), gradient)

```

# TOP HAT/WHITE HAT AND BLACK HAT

```python
args = {
	"image": "car.jpg"
}
```

```python
image = cv2.imread(args["image"])
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
rectKernel = cv2.getStructuringElement(cv2.MORPH_RECT, (13, 5))
blackhat = cv2.morphologyEx(gray, cv2.MORPH_BLACKHAT, rectKernel)
```

```python
tophat = cv2.morphologyEx(gray, cv2.MORPH_TOPHAT, rectKernel)

plt_imshow("Original", image)
plt_imshow("Blackhat", blackhat)
plt_imshow("Tophat", tophat)
```
